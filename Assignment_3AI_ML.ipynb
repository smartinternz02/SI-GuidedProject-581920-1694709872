##Find the outliers and replace them outliers


```python
import pandas as pd
import numpy as np

df.replace('NA', np.nan, inplace=True)

# Function to find outliers in a column using IQR method
def find_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = (column < lower_bound) | (column > upper_bound)

    return outliers

outliers_df = df.select_dtypes(include=[np.number]).apply(find_outliers_iqr)

print(outliers_df)
```

```python
import pandas as pd
import numpy as np

df.replace('NA', np.nan, inplace=True)

def replace_outliers_iqr(column):

    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    column[column < lower_bound] = column.median()
    column[column > upper_bound] = column.median()

    return column
numeric_columns = df.select_dtypes(include=[np.number])
for col in numeric_columns.columns:
    df[col] = replace_outliers_iqr(df[col])


```
